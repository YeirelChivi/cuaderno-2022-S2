## Ejercicio 1

usando Pseint crear un diagrama de flujo para representar el flujo, pseudocodigo y codigo de python para el algoritmo del ejercicio

- Recibira cuatro parametros
- Luego se sumaran los cuatro parametros
- del resultado de los cuatro parametros se sumaran los cuatro parametros y que se divide entre 4

### Pseudocodigo


```
Algoritmo sin_titulo
	Leer N1
	Leer N2
	Leer N3
	Leer N4
	total = (N1+N2+N3+N4)/4
	Escribir total
FinAlgoritmo

```



### Diagrama de flujo

ejercicio 1.png

### Codigo de Python


```
"""
Nombre: Ejercicio1
Entrada: valor de 4 parametros
Salida:resultado de los cuatro parametros divido entre cuatro
Restricciones: deben ser numeros enteros
"""
def Ejercicio1(num1, num2, num3, num4):
    if (isinstance( num1, int)) and (isinstance( num2, int)) and (isinstance( num3, int)) and (isinstance( num4, int)):
         
        while (num1 >= 0 and num2 >= 0 and num3 >= 0 and num4 >= 0):
            resultado = (num1+num2+num3+num4) /4
            return int(resultado)
    else:
        return"Error ingrese un numero entero"
          
```
