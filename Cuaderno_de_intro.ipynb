### Codigo de Python



```

"""
Nombre:factorial
Entrada:numero
Salida:cual numero es mayor
Restricciones:numero entero mayor a 0
"""           
def factorial(num):
    resultado = 1
    if isinstance(num, int) and (num>=0):
        while(num>0):
            resultado *= num
            num-=1
        return resultado
    else:
        return "Error en los parametros de entrada"
```
